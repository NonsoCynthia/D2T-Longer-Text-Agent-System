````python
# ============================================================
# WebNLG-style synthetic dataset generator with UnifiedModel (OpenAI o3)
# - Uses your agents.llm_model.UnifiedModel factory exactly as requested
# - Web agent pulls S,P,O from Wikidata
# - LLM agent synthesizes triples and generates descriptions
# - Excludes places and people as subjects
# - Writes XML with <synthetictripleset>, duplicated <modifiedtripleset>, and <lex>
# ============================================================

import os
import json
import time
import random
import requests
import xml.sax.saxutils as xml_utils
from pathlib import Path
from typing import List, Tuple, Dict, Any, Optional

from agents.llm_model import UnifiedModel, model_name

# ------------------------------
# LLM init per your request (OpenAI o3, temp 0.0)
# ------------------------------
provider = "openai"
conf = model_name.get(provider.lower(), {}).copy()
conf["model_name"], conf["temperature"] = "o3", 1.0
unified = UnifiedModel(provider=provider, **conf)

# ------------------------------
# Config
# ------------------------------
RANDOM_SEED = 42
random.seed(RANDOM_SEED)

N_ENTRIES = 20
MIN_TRIPLES = 10
MAX_TRIPLES = 100
OUTPUT_XML = Path("dataset2_synthetic.xml")
LANG = "en"

WEB_ATTEMPTS_PER_ENTRY = 2

# ------------------------------
# Helpers
# ------------------------------
def xescape(text: str) -> str:
    return xml_utils.escape(text, {"'": "&apos;", '"': "&quot;"})

def triple_line(tag: str, s: str, p: str, o: str) -> str:
    return f"        <{tag}>{xescape(s)} | {xescape(p)} | {xescape(o)}</{tag}>"

def slugify(x: str) -> str:
    return x.strip().replace(" ", "_")

def ensure_no_people_places_subject(triples: List[Tuple[str, str, str]]) -> List[Tuple[str, str, str]]:
    # Remove triples where the subject clearly looks like a person or place
    bad_pred_hints = {"birthPlace", "deathPlace", "born", "country", "capital_of", "population"}
    out = []
    for s, p, o in triples:
        s_l = s.lower()
        if any(tok in s_l for tok in ["city", "country", "province", "village", "person", "human", "town"]):
            continue
        if p in bad_pred_hints and s == o:
            continue
        out.append((s, p, o))
    return out


# ------------------------------
# LLM wrapper using your UnifiedModel chain
# ------------------------------
def llm_invoke(system_prompt: str, user_prompt: str) -> str:
    chain = unified.model_(agent_prompts=system_prompt)
    result = chain.invoke({"input": user_prompt})
    # langchain returns an AIMessage-like object with .content in most cases
    if hasattr(result, "content"):
        return result.content
    # or a dict with content nested
    if isinstance(result, dict) and "content" in result:
        return result["content"]
    return str(result)

# ------------------------------
# Wikidata access utilities
# ------------------------------
WIKIDATA_SEARCH = "https://www.wikidata.org/w/api.php"
WIKIDATA_SPARQL = "https://query.wikidata.org/sparql"


def wikidata_search_entity(term: str) -> Optional[str]:
    params = {
        "action": "wbsearchentities",
        "search": term,
        "language": "en",
        "format": "json",
        "type": "item",
        "limit": 1
    }
    r = requests.get(WIKIDATA_SEARCH, params=params, timeout=20)
    if r.status_code != 200:
        return None
    data = r.json()
    hits = data.get("search", [])
    if not hits:
        return None
    return hits[0]["id"]

def fetch_qid_label(qid: str) -> Optional[str]:
    params = {"action": "wbgetentities", "ids": qid, "format": "json", "languages": "en"}
    r = requests.get(WIKIDATA_SEARCH, params=params, timeout=20)
    if r.status_code != 200:
        return None
    try:
        return r.json()["entities"][qid]["labels"]["en"]["value"]
    except Exception:
        return None

def query_wikidata_triples(qid: str, limit: int = 200) -> List[Tuple[str, str, str]]:
    headers = {"Accept": "application/sparql-results+json"}
    q = SPARQL_TEMPLATE % (qid, limit)
    r = requests.get(WIKIDATA_SPARQL, params={"query": q}, headers=headers, timeout=30)
    if r.status_code != 200:
        return []
    rows = r.json().get("results", {}).get("bindings", [])
    subj_label = fetch_qid_label(qid) or qid
    subj_slug = slugify(subj_label)
    triples = []
    for row in rows:
        p = row.get("pLabel", {}).get("value", "")
        o = row.get("oLabel", {}).get("value", "")
        if not p or not o:
            continue
        triples.append((subj_slug, slugify(p), slugify(o)))
    return triples

# ------------------------------
# Agents
# ------------------------------
class WebEntityAgent:
    def fetch(self, domain: str, min_k: int, max_k: int) -> Optional[Dict[str, Any]]:
        candidates = SEEDS.get(domain, [])
        random.shuffle(candidates)
        for term in candidates:
            qid = wikidata_search_entity(term.replace("_", " "))
            if not qid:
                continue
            triples = query_wikidata_triples(qid, limit=max_k * 6)
            triples = ensure_no_people_places_subject(triples)
            if not triples:
                continue

            # Keep subject-consistent
            subj_counts = {}
            for s, _, _ in triples:
                subj_counts[s] = subj_counts.get(s, 0) + 1
            subject = max(subj_counts, key=subj_counts.get)
            triples = [(s, p, o) for (s, p, o) in triples if s == subject]

            if len(triples) >= min_k:
                random.shuffle(triples)
                triples = triples[:random.randint(min_k, min(max_k, len(triples)))]
                return {"entity": subject, "triples": triples}
        return None

class LLMTripleAgent:
    def synthesize(self, min_k: int, max_k: int) -> Dict[str, Any]:
        system_prompt = TRIPLE_AGENT_SYSTEM
        user_prompt = TRIPLE_AGENT_USER.format(min_k=min_k, max_k=max_k)
        raw = llm_invoke(system_prompt, user_prompt).strip()

        # Extract JSON
        s = raw.find("{")
        e = raw.rfind("}")
        if s != -1 and e != -1 and e > s:
            raw = raw[s:e+1]
        try:
            data = json.loads(raw)
        except Exception:
            # fallback
            entity = "Synthetic_Entity"
            triples = [(entity, "property", f"value_{i}") for i in range(min_k)]
            return {"entity": entity, "triples": triples}

        entity = slugify(data.get("entity", "Synthetic_Entity"))
        triples = []
        for t in data.get("triples", []):
            if isinstance(t, (list, tuple)) and len(t) == 3:
                s, p, o = [slugify(str(x)) for x in t]
                triples.append((s, p, o))
        triples = ensure_no_people_places_subject(triples)
        if len(triples) < min_k:
            while len(triples) < min_k:
                triples.append((entity, "property", f"value_{len(triples)}"))
        return {"entity": entity, "triples": triples[:max_k]}

class LLMTextAgent:
    def realize(self, triples: List[Tuple[str, str, str]]) -> str:
        lines = [f"{s} | {p} | {o}" for s, p, o in triples]
        system_prompt = TEXT_AGENT_SYSTEM
        user_prompt = TEXT_AGENT_USER.format(triples="\n".join(lines))
        text = llm_invoke(system_prompt, user_prompt).strip()
        return text.replace("\n", " ").strip()

# ------------------------------
# Build one entry
# ------------------------------
def build_entry(eid: int,
                web_agent: WebEntityAgent,
                llm_triples: LLMTripleAgent,
                llm_text: LLMTextAgent,
                min_triples: int,
                max_triples: int) -> str:
    domain = random.choice(DOMAINS)
    pkg = None

    for _ in range(WEB_ATTEMPTS_PER_ENTRY):
        try:
            pkg = web_agent.fetch(domain, min_triples, max_triples)
            if pkg:
                break
        except Exception:
            time.sleep(0.3)

    if not pkg:
        pkg = llm_triples.synthesize(min_triples, max_triples)

    triples = pkg["triples"]
    size = len(triples)

    try:
        lex_text = llm_text.realize(triples)
    except Exception:
        lex_text = "This entry summarizes properties of the subject based on the provided triples."

    syn_lines = ["      <synthetictripleset>"]
    for s, p, o in triples:
        syn_lines.append(triple_line("otriple", s, p, o))
    syn_lines.append("      </synthetictripleset>")

    mod_lines = ["      <modifiedtripleset>"]
    for s, p, o in triples:
        mod_lines.append(triple_line("mtriple", s, p, o))
    mod_lines.append("      </modifiedtripleset>")

    entry = []
    entry.append(f'    <entry category="{xescape(domain)}" eid="{eid}" shape="(X (X) (X))" shape-type="sibling" size="{size}">')
    entry.extend(syn_lines)
    entry.extend(mod_lines)
    entry.append(f'      <lex comment="" lid="eid{eid}_l1" lang="{LANG}">{xescape(lex_text)}</lex>')
    entry.append("    </entry>")
    return "\n".join(entry)

# ------------------------------
# Generate dataset
# ------------------------------
web_agent = WebEntityAgent()
llm_triples = LLMTripleAgent()
llm_text = LLMTextAgent()

entries_xml = []
for i in range(1, N_ENTRIES + 1):
    entries_xml.append(
        build_entry(
            eid=i,
            web_agent=web_agent,
            llm_triples=llm_triples,
            llm_text=llm_text,
            min_triples=MIN_TRIPLES,
            max_triples=MAX_TRIPLES
        )
    )

xml_text = """<?xml version="1.0" encoding="UTF-8"?>
<benchmark>
  <entries>
{entries}
  </entries>
</benchmark>
""".format(entries="\n\n".join(entries_xml))

OUTPUT_XML.write_text(xml_text, encoding="utf-8")
print(f"Wrote: {OUTPUT_XML.resolve()}")
print(xml_text.splitlines()[0], "... ok")
````
